We can see if the distribution follow a geometric law. The geometric law has parameter $p$ and the probability of seeing a particular value $k$ is given by $p(k) = (1-p)^{k-1}p$

```python
ph = float(len(data))/data.sum()print 'p = %g' % ph
```
